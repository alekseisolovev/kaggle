```python
import numpy as np 
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

for c in train.columns:
    if train[c].dtype == 'object':
        le = LabelEncoder() 
        le.fit(list(train[c].values) + list(test[c].values)) 
        train[c] = le.transform(list(train[c].values))
        test[c] = le.transform(list(test[c].values))
```

```python
train.head()
```

```python
X_train = train.drop(['y', 'ID'], axis=1)
X_test = test.drop(['ID'], axis=1)
y_train = train['y']
```

```python
from sklearn.decomposition import PCA, FastICA, TruncatedSVD
from sklearn.random_projection import GaussianRandomProjection
from sklearn.random_projection import SparseRandomProjection

pca = PCA(n_components=5)
ica = FastICA(n_components=5, max_iter=1000)
tsvd = TruncatedSVD(n_components=5)
gp = GaussianRandomProjection(n_components=5)
sp = SparseRandomProjection(n_components=5, dense_output=True)

X_train_pca = pd.DataFrame(pca.fit_transform(X_train))
X_train_ica = pd.DataFrame(ica.fit_transform(X_train))
X_train_tsvd = pd.DataFrame(tsvd.fit_transform(X_train))
X_train_gp = pd.DataFrame(gp.fit_transform(X_train))
X_train_sp = pd.DataFrame(sp.fit_transform(X_train))

X_test_pca = pd.DataFrame(pca.transform(X_test))
X_test_ica = pd.DataFrame(ica.transform(X_test))
X_test_tsvd = pd.DataFrame(tsvd.transform(X_test))
X_test_gp = pd.DataFrame(gp.transform(X_test))
X_test_sp = pd.DataFrame(sp.transform(X_test))
```

```python
X_train_pca.columns = ['pca_{}'.format(i) for i in X_train_pca.columns]
X_train_ica.columns = ['ica_{}'.format(i) for i in X_train_ica.columns]
X_train_tsvd.columns = ['tsvd_{}'.format(i) for i in X_train_tsvd.columns]
X_train_gp.columns = ['gp_{}'.format(i) for i in X_train_gp.columns]
X_train_sp.columns = ['sp_{}'.format(i) for i in X_train_sp.columns]

X_train = pd.concat((X_train, X_train_pca), axis=1)
X_train = pd.concat((X_train, X_train_ica), axis=1)
X_train = pd.concat((X_train, X_train_tsvd), axis=1)
X_train = pd.concat((X_train, X_train_gp), axis=1)
X_train = pd.concat((X_train, X_train_sp), axis=1)

X_train.head()
```

```python
X_test_pca.columns = ['pca_{}'.format(i) for i in X_test_pca.columns]
X_test_ica.columns = ['ica_{}'.format(i) for i in X_test_ica.columns]
X_test_tsvd.columns = ['tsvd_{}'.format(i) for i in X_test_tsvd.columns]
X_test_gp.columns = ['gp_{}'.format(i) for i in X_test_gp.columns]
X_test_sp.columns = ['sp_{}'.format(i) for i in X_test_sp.columns]

X_test = pd.concat((X_test, X_test_pca), axis=1)
X_test = pd.concat((X_test, X_test_ica), axis=1)
X_test = pd.concat((X_test, X_test_tsvd), axis=1)
X_test = pd.concat((X_test, X_test_gp), axis=1)
X_test = pd.concat((X_test, X_test_sp), axis=1)

X_test.head()
```

```python
from bayes_opt import BayesianOptimization
from sklearn.metrics import r2_score
import xgboost as xgb

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)

def eval_xgb(max_depth, subsample, min_child_weight, gamma, colsample_bytree):
    xgb_params = {'n_trees': 550, 
                  'eta': 0.008, 
                  'max_depth': int(max_depth), 
                  'subsample': max(min(subsample, 1), 0), 
                  'objective': 'reg:linear', 
                  'base_score': np.mean(y_train), 
                  'silent': 1, 
                  'min_child_weight': int(min_child_weight), 
                  'gamma': max(gamma, 0), 
                  'colsample_bytree': max(min(colsample_bytree, 1), 0)}
    
    scores = xgb.cv(xgb_params, dtrain, num_boost_round=1500, 
                    early_stopping_rounds=50, verbose_eval=False, 
                    feval=xgb_r2_score, maximize=True, nfold=5)['test-r2-mean'].iloc[-1]
    return scores
```

```python
params = {'min_child_weight': (1, 20), 
          'gamma': (0, 10), 
          'subsample': (0.5, 1), 
          'colsample_bytree': (0.1, 1), 
          'max_depth': (2, 15)}

dtrain = xgb.DMatrix(X_train, y_train)
dtest = xgb.DMatrix(X_test)

xgb_bayesopt = BayesianOptimization(eval_xgb, params)
xgb_bayesopt.maximize(init_points=5, n_iter=2)
```

```python
p = xgb_bayesopt.res['max']['max_params']

xgb_params = {'n_trees': 550, 
              'eta': 0.008, 
              'max_depth': int(p['max_depth']), 
              'subsample': max(min(p['subsample'], 1), 0), 
              'objective': 'reg:linear', 
              'base_score': np.mean(y_train), 
              'silent': 1, 
              'min_child_weight': int(p['min_child_weight']), 
              'gamma': max(p['gamma'], 0), 
              'colsample_bytree': max(min(p['colsample_bytree'], 1), 0)}

bst = xgb.train(xgb_params, dtrain, num_boost_round=1500, 
                verbose_eval=False, feval=xgb_r2_score, maximize=True)

print(round(r2_score(dtrain.get_label(), bst.predict(dtrain)), 2))
```

```python
y_pred = bst.predict(dtest)

output = pd.DataFrame({'id': test['ID'].astype(np.int32), 'y': y_pred})
output.to_csv('submission.csv', index=False)
```
